````python
# Python housekeeping
# Import Graph
from rdflib import Graph, Namespace, URIRef, RDF, RDFS, OWL, Literal
from rdflib.plugins.sparql import prepareQuery
from rdflib import URIRef
````

````python

# Initialize the graph
g = Graph()

# Parse the TTL files
g.parse("https://raw.githubusercontent.com/susvej/bg_ontology/refs/heads/main/bgg_main.ttl", format="ttl")

bgg = Namespace("https://raw.githubusercontent.com/susvej/bg_ontology/")
g.namespace_manager.bind("bgg", bgg, override=True)

# Print number of triples
print(f"Graph has {len(g)} triples.")

# Print first 20 triples using qname for URIs, quoted strings for literals
# This is just some random data.
for s, p, o in list(g)[:20]:
    s_str = g.qname(s) if isinstance(s, URIRef) else str(s)
    p_str = g.qname(p) if isinstance(p, URIRef) else str(p)
    if isinstance(o, URIRef):
        o_str = g.qname(o)
    elif isinstance(o, Literal):
        o_str = f'"{o}"'
    else:
        o_str = str(o)
    print(s_str, p_str, o_str)
````

````python
# Python housekeeping, but run this cell even if you have no
# interest in Python, and look at the results to explore the graph

# g.qname gives prefix:localpart rather than whole IRI
qname = g.qname

print("CLASSES:")
for s in g.subjects(RDF.type, RDFS.Class):
    print(" -", qname(s))

print("\nPROPERTIES:")
for s in g.subjects(RDF.type, RDF.Property):
    print(" -", qname(s))
for s in g.subjects(RDF.type, OWL.ObjectProperty):
    print(" -", qname(s))
for s in g.subjects(RDF.type, OWL.DatatypeProperty):
    print(" -", qname(s))

print("\nOBJECT PROPERTY RELATIONSHIPS:")
for prop in set(g.subjects(RDF.type, OWL.ObjectProperty)):
    dom = list(g.objects(prop, RDFS.domain))
    ran = list(g.objects(prop, RDFS.range))
    print(f" - {qname(prop)}")
    if dom:
        print(f"   ↳ Domain: {', '.join(qname(d) for d in dom)}")
    if ran:
        print(f"   ↳ Range:  {', '.join(qname(r) for r in ran)}")

print("\nDATATYPE PROPERTY RELATIONSHIPS:")
for prop in set(g.subjects(RDF.type, OWL.DatatypeProperty)):
    dom = list(g.objects(prop, RDFS.domain))
    ran = list(g.objects(prop, RDFS.range))
    print(f" - {qname(prop)}")
    if dom:
        print(f"   ↳ Domain: {', '.join(qname(d) for d in dom)}")
    if ran:
        print(f"   ↳ Range:  {', '.join(qname(r) for r in ran)}")

#Todo(Sus, April 5, 2025, Note that there is an error where datatype property hasMentalLoad takes a class range.
#Non-urgent fix, since there is no data of that type yet
````

````python
# Python housekeeping
# Print the results prettier.
import pandas as pd
def pretty(results):
    """
    Converts RDFLib SPARQL SELECT query results into a pandas DataFrame.

    Parameters:
        results (rdflib.query.Result): The result of a SELECT query on an RDFLib Graph.

    Returns:
        pd.DataFrame: A DataFrame with variable names as columns and query results as rows.
    """
    data = [row.asdict() for row in results]
    return pd.DataFrame(data)
````

````python
#Find the name of the highest and lowest rated games

myquery = prepareQuery("""
    PREFIX bgg: <https://raw.githubusercontent.com/susvej/bg_ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX  owl:  <http://www.w3.org/2002/07/owl#>
    PREFIX  sh: <http://www.w3.org/ns/shacl#>
    PREFIX  xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?name ?rating WHERE
{
  {
        # Subquery to get the maximum rating
        {
            SELECT (MAX(?r) AS ?targetRating) WHERE
            {
                ?g a bgg:Game ; bgg:hasGeekRating ?r .
            }
        }
        ?game a bgg:Game ;
              bgg:hasName ?name ;
              bgg:hasGeekRating ?rating .
        FILTER(?rating = ?targetRating)
  }
  UNION
  {
      # Subquery to get the minimum rating
        {
      SELECT (MIN(?r) as ?mintarget) WHERE
          {
          ?g rdf:type bgg:Game;
          bgg:hasGeekRating ?r.
          }
        }
      ?game rdf:type bgg:Game;
      bgg:hasName ?name;
      bgg:hasGeekRating ?rating.
      FILTER(?rating = ?mintarget)
    }
  }
""")
#
pretty(g.query(myquery))


````

````python
#Problem 10: Bind a rating category based on rating thresholds: <0.6 "bad", 0.6-0.8 "ok", "0.8-1.0"
myquery = prepareQuery("""
    PREFIX bgg: <https://raw.githubusercontent.com/susvej/bg_ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX  owl:  <http://www.w3.org/2002/07/owl#>
    PREFIX  sh: <http://www.w3.org/ns/shacl#>
    PREFIX  xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?n ?rat ?opinion WHERE {
  ?game rdf:type bgg:Game;
    bgg:hasName ?n;
    bgg:hasGeekRating ?rat
  BIND(
    IF(?rat <=6, "bad",
      IF(?rat >6 && ?rat <=0.8, "ok", "good"))
    as ?opinion
  )
}
ORDER BY ?rat
LIMIT 10
""")
#
pretty(g.query(myquery))
````

````python
#Problem 10: Figure out the average game play time for all games and add a rating of "shorter" if it is below that, or "longer" if it is above

myquery = prepareQuery("""
    PREFIX bgg: <https://raw.githubusercontent.com/susvej/bg_ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX  owl:  <http://www.w3.org/2002/07/owl#>
    PREFIX  sh: <http://www.w3.org/ns/shacl#>
    PREFIX  xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?name ?timerating WHERE {
  {
    SELECT (AVG(?time) as ?avgtime) WHERE {
      ?game rdf:type bgg:Game;
        bgg:hasMaxGameTime ?time
    }
  }
?game rdf:type bgg:Game;
bgg:hasName ?name;
bgg:hasMaxGameTime ?t2
BIND(
  IF(?t2 < ?avgtime, "short", "long")
  as ?timerating
)
}
""")
#
pretty(g.query(myquery))
````

````python
##See all games that expands Game1 at some level
##Note inverse query sign of ^bgg:expands+
myquery = prepareQuery("""
    PREFIX bgg: <https://raw.githubusercontent.com/susvej/bg_ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX  owl:  <http://www.w3.org/2002/07/owl#>
    PREFIX  sh: <http://www.w3.org/ns/shacl#>
    PREFIX  xsd: <http://www.w3.org/2001/XMLSchema#>
SELECT DISTINCT ?game2 WHERE {
  bgg:Game1 ^bgg:expands+ ?game2
}
LIMIT 20

""")
#
pretty(g.query(myquery))
````
